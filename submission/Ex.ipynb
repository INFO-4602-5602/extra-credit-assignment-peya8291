```python
import pandas as pd
import numpy as np
import sys
import matplotlib.pylab as plt
%matplotlib inline
import pylab as pl
from pandas import *

#read dataset
dfH= pd.read_csv("usp.csv")
dfH
#change the date type
dfH['Date Local']=pd.to_datetime(dfH['Date Local'])
#dfg= dfH.sort_values(by="Date", ascending=True)
print(dfH.head(10))

#plot the dataset line graph
ax=dfH.plot(x = "Date Local", y = ["NO2 Mean", "O3 Mean", "SO2 Mean", "CO Mean"],figsize=(20,10), grid=True,title="Pollutants Change")
ax.set_xlabel("Date", fontsize=16)
ax.set_ylabel("Value", fontsize=16)
plt.show()




```
